```python
from zipfile import ZipFile
import numpy as np
import cv2
directory = "busy/"
video = 'busy.mp4'

vidcap = cv2.VideoCapture('data/'+video)
success,image = vidcap.read()
count = 0
while success:
  cv2.imwrite("data/"+directory+str(count).zfill(6)+".jpg", image)     # save frame as JPEG file      
  success,image = vidcap.read()
  print('Read a new frame: ', success)
  count += 1

vidcap.release()

cv2.destroyAllWindows()
```
